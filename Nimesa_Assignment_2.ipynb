```python
import requests

def get_weather_forecast():
    url = "https://samples.openweathermap.org/data/2.5/forecast/hourly?q=London,us&appid=b6907d289e10d714a6e88b30761fae22"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Failed to fetch weather data.")
        return None

def get_weather_data_on_date(forecast_data, target_date):
    for data_point in forecast_data['list']:
        date_time = data_point['dt_txt']
        if target_date in date_time:
            return data_point
    return None

def get_user_input_date():
    return input("Enter the date (YYYY-MM-DD HH:MM:SS) for which you want weather data: ")

def print_weather(weather_data):
    print(f"Temperature: {weather_data['main']['temp']}°C")

def print_wind_speed(weather_data):
    print(f"Wind Speed: {weather_data['wind']['speed']} m/s")

def print_pressure(weather_data):
    print(f"Pressure: {weather_data['main']['pressure']} hPa")

def main():
    forecast_data = get_weather_forecast()
    if forecast_data is None:
        return

    while True:
        print("\nOptions:")
        print("1. Get weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")
        option = int(input("Enter your choice: "))

        if option == 1:
            target_date = get_user_input_date()
            weather_data = get_weather_data_on_date(forecast_data, target_date)
            if weather_data:
                print_weather(weather_data)
            else:
                print("Weather data not found for the given date.")
        elif option == 2:
            target_date = get_user_input_date()
            weather_data = get_weather_data_on_date(forecast_data, target_date)
            if weather_data:
                print_wind_speed(weather_data)
            else:
                print("Weather data not found for the given date.")
        elif option == 3:
            target_date = get_user_input_date()
            weather_data = get_weather_data_on_date(forecast_data, target_date)
            if weather_data:
                print_pressure(weather_data)
            else:
                print("Weather data not found for the given date.")
        elif option == 0:
            print("Exiting the program.")
            break
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()
```
